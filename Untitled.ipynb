```python
import matplotlib.pyplot as plt
def get_values(objs, feature_index, remove_outlier=False):
    feature_values = [sample['featureVector'][feature_index] for sample in objs]
    shapley_values = [sample['contributions'][0]['features'][feature_index] for sample in objs]
    if remove_outlier:
        Q1 = sorted(feature_values)[int(len(feature_values) / 4)]
        Q3 = sorted(feature_values)[int(len(feature_values) * 3 / 4)]
        min_f = Q1 - 1.5*(Q3 - Q1)
        max_f = Q1 + 1.5*(Q3 - Q1)
        new_feature_values = []
        new_shapley_values = []
        for i in range(len(feature_values)):
            if feature_values[i] >= min_f and feature_values[i] <= max_f:
                new_feature_values.append(feature_values[i])
                new_shapley_values.append(shapley_values[i])
        return new_feature_values, new_shapley_values
    return feature_values, shapley_values
    
feature_values, shapley_values = get_values(contributions, 302, remove_outlier=True)
#print(feature_values)
plt.scatter(feature_values, shapley_values)    
```

```python
from sklearn import metrics
def get_F1_from_topK(K, feature_values, shapley_values):
    min_feature = min(feature_values)
    max_feature = max(feature_values)    
    print("min_feature: {0}, max_feature: {1}".format(min_feature, max_feature))
    range_feature = (max_feature - min_feature) * 1.0
    if range_feature == 0:
        print("Skipping feature range 0")
        return -1, -1, -1, -1
    
    zipped = zip(feature_values, shapley_values)
    result = sorted(zipped, key=lambda x: x[1])[::-1]
    topK_num = int(K * len(result))
    X = []
    Y = []
    for feature, shap in result[:topK_num + 1]:
        if feature >= min_feature and feature <= max_feature:
            X.append((feature - min_feature) / range_feature)
            Y.append(1)
        
    for feature, shap in result[topK_num + 1: ]:
        if feature >= min_feature and feature <= max_feature:
            X.append((feature - min_feature) / range_feature)
            Y.append(0)
     
    print("Sample size after removing outliers: ", len(X))
    fpr, tpr, thrs = metrics.roc_curve(Y, X)
    auc = metrics.auc(fpr, tpr)
    if auc < 0.2:
        X = [1-x for x in X]
        fpr, tpr, thrs = metrics.roc_curve(Y, X)
        reverted_auc = metrics.auc(fpr, tpr)
        print("origin AUC: {0}, reverted AUC: {1}".format(auc, reverted_auc))
        auc = - reverted_auc
    else:
        print("AUC: ", metrics.auc(fpr, tpr))
    precision, recall, thresholds = metrics.precision_recall_curve(Y, X)
    print("AUPRC: ", metrics.auc(recall, precision))
#     fig, axes = plt.subplots(1, 2)
#     axes[0].plot(fpr, tpr, 'r')
#     axes[0].set_title("ROC")
#     axes[1].plot(recall, precision, 'b')
#     axes[1].set_title("PRC")
    #print(thresholds)
    max_f1 = 0
    for i in range(len(precision)):
        f1 = (2*precision[i]*recall[i])/(precision[i]+recall[i])
        if f1 > max_f1:
            max_thr = thresholds[i]
            max_f1 = f1
            
    print("Max F1: {0}, feature thr: {1}".format(max_f1, max_thr * range_feature + min_feature))
    #print('Real feature thrshold: ', max_thr * range_feature + min_feature)
    return auc, max_f1, max_thr * range_feature + min_feature, result[topK_num][1]

#get_F1_from_topK(0.15, feature_values, shapley_values)        
    
        
    
```

```python
"""
This section calculate the threshold of feature value and shapley value of training data
"""
data_size = len(contributions)
info = {} # training data
print("data size: ", data_size)
plot = True
remove_outlier = False
feature_cnt = 1
total_plot = 5
for feature_info in pos_contribution_tuples:
    _, featureId, featureName = feature_info
    print("featureId : {0}, featureName: {1}".format(featureId, featureName))
    feature_values, shapley_values = get_values(contributions, featureId, remove_outlier)
    auc, f1, thr, shap_thr = get_F1_from_topK(0.10, feature_values, shapley_values)
    if plot and feature_cnt <= total_plot and :
        plt.scatter(feature_values, shapley_values)
        plt.axvline(x=thr, color='r')
        plt.axhline(y=shap_thr, color='g')
        plt.title("xthr: {0}, y thr: {1}, top{2}: {3}".format(thr, shap_thr, feature_cnt, featureName))
        plt.savefig('Top{0}_removeOutlier_{1}.png'.format(feature_cnt, remove_outlier))
        plt.cla()
        plt.clf()
        plt.close()
    info[featureId] = {}
    info[featureId]['auc'] = auc
    info[featureId]['f1'] = f1
    info[featureId]['feature_threshold'] = thr
    info[featureId]['shapley_threshold'] = shap_thr
    feature_cnt += 1
    print()


output_file = FILE.strip('.json') + '_thresholds.json'
if remove_outlier:
    output_file = output_file.strip('.json') + '_remove_outlier.json'
with open(output_file, 'w') as f:
    json.dump(info, f)
    
```

```python
import numpy as np
def get_metrics(feature_values, shapley_values, feature_threshold, shapley_threshold, direction):
    gt = []
    pred = []
    for i in range(len(feature_values)):
        #print(shapley_values[i])
        #print(shapley_threshold)
        if shapley_values[i] > shapley_threshold:
            gt.append(1)
        else:
            gt.append(0)
        
        condition = feature_values[i] > feature_threshold if direction else feature_values[i] < feature_threshold
        if condition:
            pred.append(1)
        else:
            pred.append(0)
            
    gt = np.array(gt)
    pred = np.array(pred)
    
    tp = np.sum(gt * pred)
    fp = np.sum((gt == 0) * (pred == 1))
    fn = np.sum((gt == 1) * (pred == 0))
    tn = np.sum((gt == 0) * (pred==0))
    return tp, fp, fn, tn

def evaluations(contribution_file, threshold_file, plot=False):
    total_plot = 5
    feature_cnt = 1
    with open(threshold_file) as f:
        thresholds = json.load(f)
        remove_outlier = 'remove' in threshold_file
        print("remove outlier: ", remove_outlier)
    if remove_outlier:
        output_file = open('output_metrics_remove_outlier', 'w')
    else:
        output_file = open('output_metrics', 'w')
    #print(thresholds)
    with open(contribution_file) as f:
        contributions = json.load(f)
    for feature_info in pos_contribution_tuples:
        _, featureId, featureName = feature_info
        feature_threshold = thresholds[str(featureId)]['feature_threshold']
        shapley_threshold = thresholds[str(featureId)]['shapley_threshold']
        #print(feature_threshold)
        #print(shapley_threshold)
        train_auc = thresholds[str(featureId)]['auc']
        train_f1 = thresholds[str(featureId)]['f1']
        print("featureId : {0}, featureName: {1}, train auc: {2}, train f1: {3}".format(featureId, featureName, train_auc, train_f1), file=output_file)
        feature_values, shapley_values = get_values(contributions, featureId, remove_outlier=False)
        if plot and feature_cnt <= total_plot:
            plt.scatter(feature_values, shapley_values)
            plt.axvline(x=feature_threshold, color='r')
            plt.axhline(y=shapley_threshold, color='g')
            plt.title("xthr: {0}, y thr: {1}, top{2}: {3}".format(feature_threshold, shapley_threshold, feature_cnt, featureName))
            plt.savefig('TestSet_Top{0}_removeOutlier_{1}.png'.format(feature_cnt, remove_outlier))
            plt.cla()
            plt.clf()
            plt.close()
        feature_cnt += 1    
        data_size = len(feature_values)
        if train_auc == -1 and train_f1 == -1:
            print("Skipping invalid feature: {0}".format(featureName), file=output_file)
            continue
        tp, fp, fn, tn = get_metrics(feature_values, shapley_values, feature_threshold, shapley_threshold, train_auc > 0)
        recall = tp / data_size
        precision = tp / (tp + fp)
        f1 = 2 * recall * precision / (recall + precision)
        print("tpr: {0}, fpr: {1}, precision: {2}, f1: {3}\n".format(tp / data_size, fp / data_size, precision, f1), file=output_file)    
```
